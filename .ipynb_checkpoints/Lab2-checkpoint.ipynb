```python
#Importing basic required libraries
import numpy as np 
import pandas as pd
```

```python
#Reading the dataset
data = pd.read_csv("B:/Durham college/data.csv")
data.head()
```

```python
#General study of the dataset
data.info()
```

```python
#Removing column with null values
data = data.drop('Unnamed: 32', axis=1)
```

```python
#Verifying successful removal of null values
data.info()
```

```python
#Checking the datatypes we are dealing with
data.dtypes
```

```python
#Understanding Descriptive stats of various columns
data.describe()
```

```python
#Understanding the type of cancerous cells
B, M = data['diagnosis'].value_counts()
print('Number of Malignant : ', M)
print('Number of Benign: ', B)
print('')
print('% of cells labeled Benign', round(B / len(data) * 100, 2), '%')
print('% of cells labeled Malignant', round(M / len(data) * 100, 2), '%')
```

```python
#Visualizing the type of cells stats
import matplotlib.pyplot as plt

# Data to plot
labels = 'Benign', 'Malignant'
sizes = data['diagnosis'].value_counts()
colors = ['teal', 'crimson']
explode= [0.4,0]
# Plot
plt.pie(sizes, explode=explode, labels=labels,radius= 1280 ,colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')
fig = plt.gcf()
fig.set_size_inches(6,6)
plt.show()
```
